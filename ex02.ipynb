```python
workinglist = texts_org.split()
```

```python
workinglist
```

```python
word_list = []
for i in workinglist:
    if i[0] == '#':
        word_list.append(i)
```

```python
word_list
```

```python
clean_list = [item.strip(string.punctuation) for item in word_list]
```

```python
clean_list
```

```python
results = {}

skip  = {'a' : '', 'the' : '', 'an' : ''}
for item in clean_list:
    if item not in skip:
        try:
            results[item] += 1
        except KeyError:
            results[item] = 1
```
