```python
from selenium import webdriver
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time
%matplotlib inline

class instagram_hashtags:
    def instagram_search(self, path, url):
        self.url = url
        self.path = path
        
        driver = webdriver.Chrome(path)
        return driver.get(url)
        
    def hashtag_crawling(self, number):
        self.number = number
        
        tag_data = []
        words = []
        words_list = []
        
        for i in range(number):
            time.sleep(2)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            
            tag_data.append(soup.select('div.C4VMK > span > a'))
            
            driver.find_element_by_xpath("//a[@class = 'HBoOv coreSpriteRightPaginationArrow']").click()
        
        for x in range(number):
            for y in range(len(tag_data)):
                words += tag_data[x][y]
                
        for i in words:
            if i[0] == '#':
                words_list.append(i)
                
        return words_list
                
    def save_file_wordcloud(self, data_path, words_list):
        self.data_path = data_path
        self.words_list = words_list
        
        f = open(data_path, 'w')
        for i in range(len(words_list)):
            f.write(words_list[i] + "\n")
        f.close()
        
        text = open(data_path).read()
        
        wordcloud = WordCloud(max_font_size = 30, font_path = '/Users/markmac/Fonts/NanumSquareBold.ttf', background_color = 'white').generate(text)
        
        return wordcloud
    
    def Visualization(self, size = (40, 40), wordcloud_data, interpolation = 'bilinear', data_save = 'test'):
        self.size = size
        self.wordcloud_data = wordcloud_data
        self.interpolation = interpolation
        self.data_save = data_save
        
        plt.figure(figsize = size)
        plt.imshow(wordcloud_data, interpolation = interpolation)
        plt.axis('off') # x, yì¶• scale(ìˆ«ì) ì—†ì• ê¸°.
        plt.show()
        plt.savefig(data_save) # ì´ë¯¸ì§€ íŒŒì¼ ì €ì¥
```

```python
import time
# í•´ì‹œíƒœê·¸ë¥¼ ë¶„ì„í•˜ê¸° ìœ„í•œ Twitter ëª¨ë“ˆ 
from konlpy.tag import Twitter
# í¬ë¡¬ ë¸Œë¼ìš°ì € ì¡°ì‘ì„ ìœ„í•œ ëª¨ë“ˆ
from selenium import webdriver
# í˜ì´ì§€ ìŠ¤í¬ë¡¤ë§ì„ ìœ„í•œ ëª¨ë“ˆ
from selenium.webdriver.common.keys import Keys
# í¬ë¡¤ë§í•  url ì£¼ì†Œ
url = "https://www.instagram.com/explore/tags/python/"
# ë‹¤ìš´ë¡œë“œ ë°›ì€ driver ì£¼ì†Œ
DRIVER_DIR = '/path/to/chromedriver'
# í¬ë¡¬ ë“œë¼ì´ë²„ë¥¼ ì´ìš©í•´ ì„ì˜ë¡œ í¬ë¡¬ ë¸Œë¼ìš°ì €ë¥¼ ì‹¤í–‰ì‹œì¼œ ì¡°ì‘í•œë‹¤.
driver = webdriver.Chrome(DRIVER_DIR)
# ì•”ë¬µì ìœ¼ë¡œ ì›¹ ìì›ì„ (ìµœëŒ€) 5ì´ˆ ê¸°ë‹¤ë¦¬ê¸°
driver.implicitly_wait(5) 
# í¬ë¡¬ ë¸Œë¼ìš°ì €ê°€ ì‹¤í–‰ë˜ë©° í•´ë‹¹ urlë¡œ ì´ë™í•œë‹¤.
driver.get(url)
# ì´ ê²Œì‹œë¬¼ ìˆ˜ë¥¼ í´ë˜ìŠ¤ ì´ë¦„ìœ¼ë¡œ ì°¾ê¸°
totalCount = driver.find_element_by_class_name('_fd86t').text 
print("ì´ ê²Œì‹œë¬¼:", totalCount)
# body íƒœê·¸ë¥¼ íƒœê·¸ ì´ë¦„ìœ¼ë¡œ ì°¾ê¸°
elem = driver.find_element_by_tag_name("body") 
# alt ì†ì„±ì˜ ê°’ì„ ë‹´ì„ ë¹ˆ ë¦¬ìŠ¤íŠ¸ ì„ ì–¸
alt_list = [] 

# í˜ì´ì§€ ìŠ¤í¬ë¡¤ì„ ìœ„í•´ ì„ì‹œ ë³€ìˆ˜ ì„ ì–¸
pagedowns = 1
# ìŠ¤í¬ë¡¤ì„ 20ë²ˆ ì§„í–‰í•œë‹¤.
while pagedowns < 20:
        # PAGE_DOWN(ìŠ¤í¬ë¡¤)ì— ë”°ë¼ì„œ ê²°ê³¼ ê°’ì´ ë‹¬ë¼ì§„ë‹¤. 
        # ê¸°ë³¸ì ìœ¼ë¡œ ë¸Œë¼ìš°ì € ì¡°ì‘ì„ í†µí•´ ê°’ì„ ì–»ì–´ì˜¬ ë•ŒëŠ” ì‹¤ì œ ë¸Œë¼ìš°ì €ì— ë³´ì´ëŠ” ë¶€ë¶„ì´ì–´ì•¼ ìš”ì†Œë¥¼ ì°¾ê±°ë‚˜ íŠ¹ì • ì´ë²¤íŠ¸ë¥¼ ë°œìƒì‹œí‚¬ ìˆ˜ ìˆë‹¤.
        elem.send_keys(Keys.PAGE_DOWN)
        # í˜ì´ì§€ ìŠ¤í¬ë¡¤ íƒ€ì´ë°ì„ ë§ì¶”ê¸° ìœ„í•´ sleep
        time.sleep(1)
        # ë¸Œë¼ìš°ì €ì— ë³´ì´ëŠ” ëª¨ë“  img íƒœê·¸ë¥¼ css ì„ íƒì ë¬¸ë²•ìœ¼ë¡œ ì°¾ëŠ”ë‹¤.
        img = driver.find_elements_by_css_selector('div.KL4Bh > img')
        # ìœ„ì—ì„œ ì„ ì–¸í•œ alt_list ë¦¬ìŠ¤íŠ¸ì— alt ì†ì„±ì˜ ê°’ì„ í• ë‹¹í•œë‹¤.
        for i in img:
            alt_list.append(i.get_attribute('alt'))
        pagedowns += 1

# ê°’ì˜ ì¤‘ë³µì„ ë°©ì§€ë¥¼ ë¦¬ìŠ¤íŠ¸ setìœ¼ë¡œ ë³€í™˜í›„ ë¦¬ìŠ¤íŠ¸ë¡œ ì¬í• ë‹¹
alt_list = list(set(alt_list))

# í‚¤:í•´ì‹œíƒœê·¸, ê°’:íšŸìˆ˜ í˜•ì‹ìœ¼ë¡œ ì €ì¥í•˜ê¸° ìœ„í•œ ë¹ˆ ë”•ì…”ë„ˆë¦¬ ì„ ì–¸
dict_data = {}
# alt ì†ì„±ì˜ ê°’ì¸ ì œëª©ê³¼ í•´ì‹œíƒœê·¸ ì¤‘ í•´ì‹œíƒœê·¸ ë§Œì„ ê°€ì ¸ì˜¤ê¸° ìœ„í•œ Tiwitter ê°ì²´ ìƒì„±
tw = Twitter()
# alt_listì— ë‹´ê¸´ ê°’ì˜ í¬ê¸°ë§Œí¼ ë°˜ë³µí•œë‹¤.
for alt in alt_list:
    # pos ë©”ì„œë“œë¥¼ í†µí•´ alt ì†ì„±ì˜ ëª¨ë“  í•´ì‹œíƒœê·¸ì˜ ê°’ì„ (ê°’, í’ˆì‚¬) í˜•íƒœì˜ íŠœí”Œì„ ìš”ì†Œë¡œ ê°–ëŠ” ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜í•œë‹¤.
    temp = tw.pos(alt, norm = True)
    # ë¦¬ìŠ¤íŠ¸ì˜ í¬ê¸°ë§Œí¼ ë°˜ë³µí•œë‹¤.
    for data in temp:
        # í’ˆì‚¬ê°€ ë§Œì•½ í•´ì‹œíƒœê·¸ì´ë©´
        if data[1] == "Hashtag":
            # ê²°ê³¼ ê°’ì„ ì €ì¥í•  ë”•ì…”ë„ˆë¦¬ì— ê°’ì´ ìˆëŠ”ì§€ í™•ì¸í•˜ê³  ì—†ë‹¤ë©´ ìƒˆë¡œì´ í‚¤ë¥¼ ì¶”ê°€í•˜ê³  0, ìˆë‹¤ë©´ ê¸°ì¡´ í‚¤ì— 1ì„ ë”í•´ì¤€ë‹¤.
            if not (data[0] in dict_data):
                dict_data[data[0]] = 0
            dict_data[data[0]] += 1

# ë”•ì…”ë„ˆë¦¬ë¥¼ íšŸìˆ˜ë¥¼ ê°€ì§€ê³  ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•œë‹¤.
keys = sorted(dict_data.items(), key = lambda x:x[1], reverse = True)
# 1~15ìœ„ ê¹Œì§€ì˜ í‚¤:ê°’ì„ ì¶œë ¥í•œë‹¤.
for k, v in keys[:15]:
    print("{}({})".format(k, v))

# ë“œë¼ì´ë²„ë¥¼ ì¢…ë£Œí•œë‹¤.
driver.close()    
```
