```python
from bs4 import BeautifulSoup
from selenium import webdriver
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time
%matplotlib inline

class instagram_Hashtags:
    def is_Accounts_Crawling(self, is_pass, search_word, is_Accounts_num, path = '/Users/markmac/Downloads/chromedriver', ):
        #self.is_id = is_id
        self.is_pass = is_pass
        self.search_word = search_word
        self.is_Accounts_num = is_Accounts_num
        self.path = path
        
        is_id = []
        
        driver = webdriver.Chrome(path)
        driver.get('https://www.instagram.com/')
        
        driver.find_element_by_xpath("//button[@class = '_0mzm- sqdOP  L3NKy       ']").click()
        
        driver.find_element_by_name('email').send_keys('ebfkdrhqnzo123@gmail.com')
        driver.find_element_by_name('pass').send_keys(is_pass)
        
        driver.find_element_by_xpath("//button[@class = '_42ft _4jy0 _52e0 _4jy6 _4jy1 selected _51sy']").click()
        
        time.sleep(3)
        driver.find_element_by_xpath("//button[@class = 'aOOlW  bIiDR  ']").click()
        
        driver.find_element_by_xpath("//div[@class = 'eyXLr wUAXj ']").click()
        driver.find_element_by_xpath("//input[@class = 'XTCLo x3qfX ']").send_keys(search_word)
        time.sleep(1)
        driver.find_element_by_xpath("//div[@class = 'z556c']").click()
        
        time.sleep(2)
        driver.find_element_by_xpath("//div[@class = '_9AhH0']").click()
        
        for i in range(is_Accounts_num):
            time.sleep(1)
            html = driver.page_source
            dom = BeautifulSoup(html, 'html.parser')
        
            term = dom.select('h2.BrX75 a')
        
            for j in term:
                is_id.append(j.text)
                
            driver.find_element_by_xpath("//a[@class = 'HBoOv coreSpriteRightPaginationArrow']").click()
            
        driver.close()
        
        return is_id
    
    def is_Hashtags_Crawling(self, is_id, path = '/Users/markmac/Downloads/chromedriver'):
        self.is_id = is_id
        self.path = path
        
        tag_data = []
        words = []
        words_list = []
        
        for i in range(len(is_id)):
            driver = webdriver.Chrome(path)
            driver.get('https://www.instagram.com/' + is_id[i])
            
            driver.find_element_by_xpath("//div[@class = '_9AhH0']").click()
            
            while True:
                try:
                    time.sleep(3)
                    html = driver.page_source
                    dom = BeautifulSoup(html, 'html.parser')
                    
                    tag_data += dom.select('div.C4VMK > span > a')
                    
                    driver.find_element_by_xpath("//a[@class = 'HBoOv coreSpriteRightPaginationArrow']").click()
                except:
                    driver.close()
                    break
                    
            time.sleep(1)
        
        for j in tag_data:
            words.append(j.text)
        
        for x in words:
            if x[0] == '#':
                words_list.append(x)
                
        return words_list
    
    def is_Hashtags_save(self, save_file_path, words_list,  max_font_size = 30, font_path = '/Users/markmac/Fonts/NanumSquareBold.ttf', background_color = 'white'):
        self.save_file_path = save_file_path
        self.words_list = words_list
        self.max_font_size = max_font_size
        self.font_path = font_path
        self.background_color = background_color
        
        f = open(save_file_path, 'w')
        
        for i in range(len(words_list)):
            f.write(words_list[i] + '\n')
        
        f.close()
        
        text = open(save_file_path).read()
        
        wordcloud = WordCloud(max_font_size, font_path, background_color).generate(text)
        
        return wordcloud
        
    def is_Wordcloud_Plot(self, wordcloud_data, save_file_name, figsize = (40, 40), interpolation = 'bilinear'):
        self.wordcloud_data = wordcloud_data
        self.save_file_name = save_file_name
        self.figsize = figsize
        self.interpolation = interpolation
        
        plt.figure(figsize)
        plt.imshow(wordcloud, interpolation)
        plt.axis('off')
        plt.savefig(save_file_name)
```
